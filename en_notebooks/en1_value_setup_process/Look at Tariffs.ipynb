```python
rates = [c for c in df.columns if 'rate' in c and not 'fit' in c ]
rates              
```

```python
for tid in df.index:
    named_rates = [c for c in rates if
                               not np.isnan(df.loc[tid, c]) and 
                               c.replace('rate', 'name') in df.columns]
                       
    solar_rates = [c for c in named_rates if 
                           'solar' in df.loc[tid, c.replace('rate', 'name')]
                        or 'Solar' in df.loc[tid, c.replace('rate', 'name')]
                               ]
    discounted_rates = [c for c in rates if c not in solar_rates]
    print(tid, discounted_rates)
```

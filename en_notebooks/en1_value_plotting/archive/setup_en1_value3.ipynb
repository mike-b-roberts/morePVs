```python
# Set up all variables
studybase = '_value3'
sites=['A','B','C','D','E','F','G','H','I','J']
pvs = {'A':['max'],
      'B':['max'],
      'C':['max'],
      'D':['max'],
      'E':['max'],
      'F':['max','0_5kw','1_0kw'],
      'G':['max','0_5kw','1_0kw','1_5kw'],
      'H':['max','0_5kw','1_0kw','1_5kw','2_0kw','2_5kw'],
      'I':['max','0_5kw','1_0kw'],
      'J':['max','0_5kw','1_0kw','1_5kw','2_0kw','2_5kw']
      }
pv_base = 'vb_pv\\site_'
btm_i = ['btm_i','btm_icp']
btm_s = ['btm_s_c', 'btm_s_u']
btms_capexes= ['allume_1', 'upfront_1']
en_capexes= ['capex_4','capex_5','capex_6']
parents = ['EA310_12c_FIT8', 'EA310_12c_FIT12', 'EA310_12c',
           'EA310_10c_FIT8', 'EA310_10c_FIT12', 'EA310_10c',
         'EA310_8c_FIT8', 'EA310_8c_FIT12','EA310_8c']
terms =[5,10,15,20,25]
rates = [0.06]
btm_tariffs = ['EASO_TOU_15pc_FIT8','EASO_TOU_15pc_FIT12']
bau_tariff= btm_tariffs[0]
net_tariff='EA310'
```

```python
# Separate `.csv` for each site:
for site in sites:
    fn = 'study_site'+site+studybase +'.csv'
    file = os.path.join(i_path,fn)
    df = pd.DataFrame()
    df.index.name='scenario'
    lf = 'h_all_site_'+site
    
    arr ='bau'
    s=0
    df.loc[s,'pv_filename'] = ''
    df.loc[s,'load_folder'] = lf
    df.loc[s,'arrangement'] = arr
    df.loc[s,'pv_cap_id'] = ''
    df.loc[s,'cp'] = bau_tariff
    df.loc[s,'all_residents'] = bau_tariff
    df.loc[s,'parent'] = 'TIDNULL'
    df.loc[s,'network_tariff'] = net_tariff
    df.loc[s,'en_capex_id'] = ''
    df.loc[s,'a_term'] = terms[0]
    df.loc[s,'a_rate'] = rates[0]
    
    arr= 'cp_only'
    parent ='TIDNULL'
    for pv in pvs[site]:     
        for tariff in btm_tariffs:
            for term in terms:
                for rate in rates:
                    s += 1
                    df.loc[s,'pv_filename'] = pv_base + site+ '_' + pv +'.csv'
                    df.loc[s,'load_folder'] = lf
                    df.loc[s,'arrangement'] = arr
                    df.loc[s,'pv_cap_id'] = 'site_'+site +'_'+ pv
                    df.loc[s,'cp'] = tariff
                    df.loc[s,'all_residents'] = tariff
                    df.loc[s,'parent'] = 'TIDNULL'
                    df.loc[s,'network_tariff'] = net_tariff
                    df.loc[s,'en_capex_id'] = ''
                    df.loc[s,'a_term'] = term
                    df.loc[s,'a_rate'] = rate

    for arr in btm_s:
        parent ='TIDNULL'
        for pv in pvs[site]:     
            for tariff in btm_tariffs:
                for capex in btms_capexes:
                    for term in terms:
                        for rate in rates:
                            s += 1
                            df.loc[s,'pv_filename'] = pv_base + site+ '_' + pv +'.csv'
                            df.loc[s,'load_folder'] = lf
                            df.loc[s,'arrangement'] = arr
                            df.loc[s,'pv_cap_id'] = 'site_'+site +'_'+ pv
                            df.loc[s,'cp'] = tariff
                            df.loc[s,'all_residents'] = tariff
                            df.loc[s,'parent'] = 'TIDNULL'
                            df.loc[s,'network_tariff'] = net_tariff
                            df.loc[s,'en_capex_id'] = capex
                            df.loc[s,'a_term'] = term
                            df.loc[s,'a_rate'] = rate
                        
    for arr in btm_i:
        parent ='TIDNULL'
        for pv in pvs[site]:     
            for tariff in btm_tariffs:
                capex ='' 
                for term in terms:
                    for rate in rates:
                        s += 1
                        df.loc[s,'pv_filename'] = pv_base + site+ '_' + pv +'.csv'
                        df.loc[s,'load_folder'] = lf
                        df.loc[s,'arrangement'] = arr
                        df.loc[s,'pv_cap_id'] = 'site_'+site +'_'+ pv
                        df.loc[s,'cp'] = tariff
                        df.loc[s,'all_residents'] = tariff
                        df.loc[s,'parent'] = parent
                        df.loc[s,'network_tariff'] = net_tariff
                        df.loc[s,'en_capex_id'] = capex
                        df.loc[s,'a_term'] = term
                        df.loc[s,'a_rate'] = rate                    
                                       
    arr ='en_pv'
    tariff = 'TIDNULL'
    for pv in pvs[site]:
        for parent in parents:           
            for capex in en_capexes:
                for term in terms:
                    for rate in rates:
                        s += 1
                        df.loc[s,'pv_filename'] = pv_base + site+ '_' + pv +'.csv'
                        df.loc[s,'load_folder'] = lf
                        df.loc[s,'arrangement'] = arr
                        df.loc[s,'pv_cap_id'] = 'site_'+site +'_'+ pv
                        df.loc[s,'cp'] = tariff
                        df.loc[s,'all_residents'] = tariff
                        df.loc[s,'parent'] = parent
                        df.loc[s,'network_tariff'] = net_tariff
                        df.loc[s,'en_capex_id'] = capex
                        df.loc[s,'a_term'] = term
                        df.loc[s,'a_rate'] = rate 
                
    arr ='en'
    tariff = 'TIDNULL'
    for parent in parents:        
        for capex in en_capexes:
            for term in terms:
                for rate in rates:
                    s += 1
                    df.loc[s,'pv_filename'] = ''
                    df.loc[s,'load_folder'] = lf
                    df.loc[s,'arrangement'] = arr
                    df.loc[s,'pv_cap_id'] = ''
                    df.loc[s,'cp'] = tariff
                    df.loc[s,'all_residents'] = tariff
                    df.loc[s,'parent'] = parent
                    df.loc[s,'network_tariff'] = net_tariff
                    df.loc[s,'en_capex_id'] = capex
                    df.loc[s,'a_term'] = term
                    df.loc[s,'a_rate'] = rate            
    
    um.df_to_csv(df,file)
                
                
```

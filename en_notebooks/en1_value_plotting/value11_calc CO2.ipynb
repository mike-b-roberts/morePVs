```python
# Look at npv calculations and calc alternate version with avoided emissions
for s in df.index:
    df.loc[s,'check_npv_whole_building'] = -sum(df.loc[s,'total$_building_costs_mean'] /(12*(1 + df.loc[s,'a_rate'] / 12) ** t)
                                      for t in np.arange(1, 12 * df.loc[s,'a_term']))

# Checks out
```

```python
# redo npv calc with CO2 price included as annual cost:
df['inclusive_total_cost']= df['total$_building_costs_mean'] - df['carbon_offset_$']
for s in df.index:
    df.loc[s,'carbon_inclusive_npv_whole_building'] = -sum(df.loc[s,'inclusive_total_cost'] /(12*(1 + df.loc[s,'a_rate'] / 12) ** t)
                                      for t in np.arange(1, 12 * df.loc[s,'a_term']))
```

```python
df['npv_cf_bau_co2']= df['carbon_inclusive_npv_whole_building'].divide(df['number_of_households'])-df['bau_npv']
```

```python
df['npv_cf_bau_pa'] = df['npv_cf_bau'].divide(df['a_term'])
df['npv_cf_bau_pa_co2'] = df['npv_cf_bau_co2'].divide(df['a_term'])
```

```python
for s in df.index:
    site = df.loc[s,'site']
    bau_scenario = df.loc[s,'bau_scenario']
    df.loc[s,'annual_generation'] = df.loc[bau_scenario,'import_kWh_mean'] - df.loc[s,'import_kWh_mean'] \
                                    - df.loc[bau_scenario,'export_kWh_mean'] + df.loc[s,'export_kWh_mean']
```

```python
carbon_price = 79.53/1000 
# 79.53 AUD / tonne 
df['carbon_offset_$']= df['annual_generation']*gh_factor*carbon_price
df['annual_carbon_offset_$_unit']= df['carbon_offset_$'].divide(df['number_of_households'])
df['total_carbon_offset_$_unit']= df['annual_carbon_offset_$_unit'].multiply(df['a_term'])
df['CO2_%_BAU'] = df['total_carbon_offset_$_unit'].divide(df['npv_cf_bau']) *100
```
